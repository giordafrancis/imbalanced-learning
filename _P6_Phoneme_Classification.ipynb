## Phoneme Dataset

In this project,  we will use a standard imbalanced machine learning dataset referred to asthePhonemedataset.  This dataset is credited to the ESPRIT (European Strategic Programon Research in Information Technology) project titled ROARS(Robust Analytical Speech Recognition System) and described in progress reports and technical reports from that project.

*The goal of the ROARS project is to increase the robustness of an existing analyticalspeech recognition system (i.e., one using knowledge about syllables, phonemes andphonetic features), and to use it as part of a speech understanding system withconnected words and dialogue capability.  This system will be evaluated for a specificapplication in two European languages* ESPRIT: The European Strategic Programme for Research and development in InformationTechnology.

The goal of the dataset was to distinguish between nasal and oral vowels.  Vowel sounds werespoken and recorded to digital files.  Then audio features were automatically extracted fromeach sound.

There are two classes for the two types of sounds; they are:
- Class 0:  Nasal Vowels (majority class).
- Class 1:  Oral Vowels (minority class)


### Explore dataset

The Phoneme dataset is a widely used standard machine learning dataset, used to explore anddemonstrate many techniques designed specifically for imbalanced classification.  One exampleis the popular SMOTE data oversampling technique. 

```python
import pandas as pd
from pandas.plotting import scatter_matrix
import numpy as np
import seaborn as sns
from collections import Counter
from matplotlib import pyplot

sns.set()
```

```python
#url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/phoneme.csv"
data = pd.read_csv("data/phoneme.csv", index_col=None)
```

```python
target = data.values[:, -1]
print(data.shape)
counter = Counter(target)
for k, v in counter.items():
    pc = v / target.shape[0]
    print(f"Class:{k},Count:{v}, Percentage={pc:.3f}")
```
